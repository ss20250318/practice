### 5.1.3 平均二乗誤差

### 5.1.4 パラメータを求める（勾配法）

```python
# リスト 5-1-(6)
# 平均二乗誤差(MSE)関数 ----------
def mse_line(x, t, w):
    y = w[0] * x + w[1]          # 式5-4、yを求める　直線の式
    mse = np.mean((y - t) ** 2)  # 式5-3、平均二乗誤差
    return mse

# 各w0、w1で平均二乗誤差(MSE)を計算 ----------
w0_n, w1_n = 100, 100                   # グラフ表示の解像度
w0_min, w0_max = -25, 25                # w0の計算範囲
w1_min, w1_max = 120, 170               # w1の計算範囲
w0 = np.linspace(w0_min, w0_max, w0_n)  # w0を準備 -25~25を100分割
w1 = np.linspace(w1_min, w1_max, w1_n)  # w1を準備 120~170を100分割
J = np.zeros((w1_n, w0_n))              # MSEを入れる配列Jを準備 100行100列の行列に0を格納
# グリッド状の点(w0, w1)に対してJを計算
for i0 in range(w0_n): #0-i0-99
    for i1 in range(w1_n): #0-i1-99
        w = np.array([w0[i0], w1[i1]]) #座標を1次元配列で表現
        J[i1, i0] = mse_line(X, T, w)
ww0, ww1 = np.meshgrid(w0, w1)          # グリッド座標の作成

# グラフ描画 ----------
plt.figure(figsize=(9.5, 4))
plt.subplots_adjust(wspace=0.5) #グラフ間の横幅
# サーフェス表示
ax = plt.subplot(1, 2, 1, projection="3d") #1行2列の１つ目
ax.plot_surface(
    ww0, ww1, J,
    rstride=10, cstride=10, alpha=0.3, color="blue", edgecolor="black",
)  #rstride:何行おきに線を引くか, cstride:何列おきに線を引くか
ax.set_xticks([-20, 0, 20])     # x軸の目盛り指定
ax.set_yticks([120, 140, 160])  # y軸の目盛り指定
ax.view_init(20, -60)           # グラフの向きの指定
# 等高線表示
plt.subplot(1, 2, 2)
cont = plt.contour(
    ww0, ww1, J, colors="black",
    levels=[100, 1000, 10000, 100000],  # 描く等高線の値を指定
)
cont.clabel(fmt="%d", fontsize=8)
plt.grid()
plt.show()
```

```python
# リスト 5-1-(7)
# 平均二乗誤差(MSE)の勾配 ----------
def dmse_line(x, t, w):
    y = w[0] * x + w[1]
    d_w0 = 2 * np.mean((y - t) * x)  # 式5-9 平均二乗誤差をw0で偏微分した値
    d_w1 = 2 * np.mean(y - t)        # 式5-10　平均二乗誤差をw1で偏微分した値
    return d_w0, d_w1
```

```python
# リスト 5-1-(8)
w = np.array([10, 165])
d_w = dmse_line(X, T, w)
print(np.round(d_w, 2))
```

```python
# リスト 5-1-(9)
# 勾配法 ----------
def fit_line_num(x, t, w_init):
    # パラメータ
    alpha = 0.001     # 学習率
    tau_max = 100000  # 繰り返しの最大数
    eps = 0.1         # 繰り返し計算を終了するための閾値
    # 勾配法
    w = np.zeros((tau_max + 1, 2))  # 変化するwの履歴を入れる配列, 100001行2列の行列の0を格納
    w[0, :] = w_init                # wの初期値をセット
    for tau in range(tau_max): #0-tau-100000
        dmse = dmse_line(x, t, w[tau, :]) #平均二乗誤差を偏微分した値を格納(w0,w1)
        w[tau + 1, 0] = w[tau, 0] - alpha * dmse[0]  # 式5-6 w0の勾配を更新
        w[tau + 1, 1] = w[tau, 1] - alpha * dmse[1]  # 式5-7 w1の勾配を更新
        if max(np.absolute(dmse)) < eps:  # 終了判定 dmse < 0.1を満たすときに終了
            break  # tau のループから抜ける
    w_final = w[tau + 1, :]   # 最終的に得られたw
    w_hist = w[: tau + 2, :]  # wの履歴で更新した分を抜き出す(w_initを含む)
    return w_final, dmse, w_hist

# メイン ----------
# 勾配法でwを計算
w_init = np.array([10.0, 165.0])               # wの初期値
w, dmse, w_history = fit_line_num(X, T, w_init)  # wを計算
mse = mse_line(X, T, w)                        # MSE(平均二乗誤差)を計算
# 結果表示
print(f"繰り返し回数 {w_history.shape[0]-1}") #w_historyの行数-1
print(f"w0 = {w[0]:.6f}, w1 = {w[1]:.6f}")
print(f"dMSE = [{dmse[0]:.6f}, {dmse[1]:.6f}]") #w0, w1の勾配を表示
print(f"MSE = {mse:.6f}") #二乗平均誤差の表示
print(f"SD = {np.sqrt(mse):.6f} cm") #標準偏差

# グラフ描画 ----------
plt.figure(ﬁgsize=(4, 4))
# 等高線表示
cont = plt.contour(
    ww0, ww1, J,     # リスト5-1-(6)で作成済
    colors="black", levels=[100, 1000, 10000, 100000],
)
cont.clabel(fmt="%1.0f", fontsize=8)
# 等高線の上に、過去のすべてのwをプロット
plt.plot(  
    w_history[:, 0], w_history[:, 1], "gray",
    marker=".",                        # マーカーの形
    markersize=10,                     # マーカーの大きさ
    markeredgecolor="cornflowerblue",  # マーカーの輪郭の色
)
plt.grid()
plt.show()
```

### 5.1.5 直線モデルパラメータの解析解

```python
# リスト 5-1-(11)
# 解析解 ----------
def fit_line(x, t):
    mx = np.mean(x)        # <x>
    mt = np.mean(t)        # <t>
    mtx = np.mean(t * x)   # <tx>
    mxx = np.mean(x * x)   # <x^2>
    w0 = (mtx - mt * mx) / (mxx - mx ** 2)  # 式5-20
    w1 = mt - w0 * mx                       # 式5-21
    w = np.array([w0, w1])
    return w

# メイン ----------
w = fit_line(X, T)       # 解析解でwを計算
mse = mse_line(X, T, w)  # MSE(平均二乗誤差)を計算
# 結果表示
print(f"w0 = {w[0]:.2f}, w1 = {w[1]:.2f}") #w0,w1の表示
print(f"MSE = {mse:.2f}") #平均事情誤差の表示
print(f"SD = {np.sqrt(mse):.2f} cm") #標準偏差の表示

# グラフ描画 ----------
plt.figure(figsize=(4, 4))
show_line(w)
plt.plot(
    X, T, "cornflowerblue",
    marker="o", linestyle="None", markeredgecolor="black",
)
plt.xlim(X_min, X_max)
plt.grid()
plt.show()
```

### 5.2.3 面モデル

```python
# リスト 5-2-(4)
# 面の表示 ----------
def show_plane(ax, w):
    # 表示データの計算
    x0_n, x1_n = 5, 5
    x0 = np.linspace(X0_min, X0_max, x0_n) #5~30を5分割
    x1 = np.linspace(X1_min, X1_max, x1_n) #40~75を5分割
    xx0, xx1 = np.meshgrid(x0, x1)  # グリッド座標の作成
    y = w[0] * xx0 + w[1] * xx1 + w[2]  # (A) 式5-28 身長を求めるモデル式に面を作るために分割したxx0,xx1を代入した
    # サーフェス表示
    ax.plot_surface(
        xx0, xx1, y,
        rstride=1, cstride=1, alpha=0.3, color="blue", edgecolor="black",
    ) #rstride:何行おきに線を引くか, cstride:何列おきに線を引くか

# 面の平均二乗誤差(MSE)関数 ----------
def mse_plane(x0, x1, t, w):
    y = w[0] * x0 + w[1] * x1 + w[2]  # (A) 式5-28
    mse = np.mean((y - t) ** 2) #(モデル式で求めた身長の予測値-実測値)^2の平均
    return mse

# メイン ----------
w = np.array([1.5, 1, 90])
mse = mse_plane(X0, X1, T, w)  # MSEを計算
# 結果表示
print(f"SD = {np.sqrt(mse):.2f} cm") #標準偏差

# グラフ描画 ----------
plt.figure(figsize=(6, 5))
ax = plt.subplot(projection="3d")
show_plane(ax, w) #面の表示
show_data2d(ax, X0, X1, T) #2次元データの表示
plt.show()
```

### 5.2.4 面モデルパラメータの解析解

```python
# リスト 5-2-(5)
# 解析解 ----------平均二乗誤差の式をw0,w1,w2で偏微分した結果
def fit_plane(x0, x1, t):  
    c_tx0 = np.mean(t * x0) - np.mean(t) * np.mean(x0)     # cov(t, x0)
    c_tx1 = np.mean(t * x1) - np.mean(t) * np.mean(x1)     # cov(t, x1)
    c_x0x1 = np.mean(x0 * x1) - np.mean(x0) * np.mean(x1)  # cov(x0, x1)
    v_x0 = np.var(x0)                                      # var(x0)
    v_x1 = np.var(x1)                                      # var(x1)
    # 式5-34
    w0 = (c_tx1 * c_x0x1 - v_x1 * c_tx0) / (c_x0x1 ** 2 - v_x0 * v_x1)
    # 式5-35
    w1 = (c_tx0 * c_x0x1 - v_x0 * c_tx1) / (c_x0x1 ** 2 - v_x0 * v_x1)
    # 式5-36
    w2 = -w0 * np.mean(x0) - w1 * np.mean(x1) + np.mean(t)
    w = np.array([w0, w1, w2])
    return w

# メイン ----------
w = fit_plane(X0, X1, T)       # wを計算
mse = mse_plane(X0, X1, T, w)  # MSE(平均二乗誤差)を計算
# 結果表示
print(f"w0 = {w[0]:.2f}, w1 = {w[1]:.2f}, w2 = {w[2]:.2f}")
print(f"SD = {np.sqrt(mse):.2f} cm")

# グラフ描画 ----------
plt.figure(figsize=(6, 5))
ax = plt.subplot(projection="3d")
show_plane(ax, w) #面の表示
show_data2d(ax, X0, X1, T) #2次元データの表示
plt.show()
```

## 5.4 線形基底関数モデル

```python
%reset
```

```python
# リスト 5-3-(1)
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# データのロード ----------
data = np.load("ch5_data.npz")
X = data["X"] #年齢のデータ
X_min = 0
X_max = data["X_max"] #X_max=30
N = data["N"] #N=16 データ数
T = data["T"] #身長のデータ
```

```python
# リスト 5-3-(2)
# ガウス関数 ----------
def gauss(x, mu, s):
    y = np.exp(-((x - mu) ** 2) / (2 * s ** 2))  # 式5-64　ガウス関数の式
    return y
```

```python
# リスト 5-3-(3)
# メイン ----------
M = 4                                # ガウス関数の数
mu = np.linspace(5, 30, M)           # 平均パラメータ:5~30を４つに分割する
s = mu[1] - mu[0]                    # (A) 標準偏差パラメータはガウス関数の中心間の距離にする
xb = np.linspace(X_min, X_max, 100)  # 0~30を100分割する
y = np.zeros((M, 100))  # M個のガウス関数の値を入れるyを準備:4行100列の行列に0を格納
for j in range(M): # 0-j-3
    y[j, :] = gauss(xb, mu[j], s)    # ガウス関数 行ごとに中心が異なるガウス関数の値が格納されている

# グラフ描画 ----------
plt.figure(figsize=(4, 4))
for j in range(M): # 0-j-3
    plt.plot(xb, y[j, :], "gray", linewidth=3)
plt.xlim(X_min, X_max) # 0~30
plt.ylim(0, 1.2)
plt.grid()
plt.show() # グラフの左側からφ0(x)~φ4(x)
```

```python
# リスト 5-3-(4)
# 線形基底関数モデル ----------
def gauss_func(w, x): 
    m = len(w) - 1        # ガウス関数の数:パラメータの数から切片を引いている
    mu = np.linspace(5, 30, m) # 5~30をm分割する
    s = mu[1] - mu[0] # 標準偏差パラメータはガウス関数の中心間の距離にする
    # xと同じサイズで要素が0のndarray型を作成
    y = np.zeros_like(x) # 要素数がxの一次元配列に0を格納する
    # ここでは式5-66ではなく式5-65で実装
    for j in range(m): # 0-j-(m-1)
        y = y + w[j] * gauss(x, mu[j], s) # w[j]*ガウス関数の結果(φ[j])の和
    y = y + w[m]  # phiを掛けないパラメータ(切片)を最後に加える
    return y　# modelの予測値
```

```python
# リスト 5-3-(5)
# 線形基底関数モデルの平均二乗誤差(MSE) ----------
def mse_gauss_func(x, t, w):
    y = gauss_func(w, x)
    mse = np.mean((y - t) ** 2)
    return mse
```

```python
# リスト 5-3-(6)
# 線形基底関数モデルの厳密解 ----------
def fit_gauss_func(x, t, m):
    mu = np.linspace(5, 30, m) # 5~30をm分割する (ex) 5~30を4分割する
    s = mu[1] - mu[0] # 標準偏差パラメータはガウス関数の中心間の距離にする
    n = x.shape[0] # xの行数 (ex) n=16
    # 式5-69 の計画行列phiを作成
    phi = np.ones((n, m + 1))  # 要素が1のn行(m+1)列の行列 (ex) 16行4列
    for j in range(m): # 0-j-(m-1) (ex) 0-j-3
        phi[:, j] = gauss(x, mu[j], s) # xを基底関数に置き換えて保存する(今回はガウス関数に置き換える)
    # 式5-68 で厳密解のwを計算
    w = np.linalg.inv(phi.T @ phi) @ phi.T @ t # np.linalg.inv:逆行列
    return w
```

```python
# リスト 5-3-(7)
# ガウス基底関数表示 ----------
def show_gauss_func(w):
    x = np.linspace(X_min, X_max, 100) # 0~30を100分割する
    y = gauss_func(w, x) # gauss_func():モデルの予測値yを求める
    plt.plot(x, y, "gray", linewidth=4) # 図に示す

# メイン ----------
M = 4                          # ガウス関数の数
w = fit_gauss_func(X, T, M)    # fit_gauss_funx():wを計算
mse = mse_gauss_func(X, T, w)  # MSE(平均二乗誤差)を計算
# 結果表示
print("w = ", np.round(w, 2)) # パラメータを小数第2位まで表示
print(f"SD = {np.sqrt(mse):.2f} cm") # 標準偏差を小数第2位まで表示

# グラフ描画 ----------
plt.figure(figsize=(4, 4))
show_gauss_func(w) # モデルの予測値をグラフに表示
plt.plot( # 実測値をグラフに表示
    X, T, "cornflowerblue", 
    marker="o", linestyle="None", markeredgecolor="black",
) 
plt.xlim(X_min, X_max) # 0~30
plt.grid()
plt.show()
```

```python
# リスト 5-3-(12)
# K 分割交差検証 ----------
def kfold_gauss_func(x, t, m, k):
    n = x.shape[0] #xの行数 (ex) n=16
    mse_train = np.zeros(k) #要素数kの一次元配列で要素は0 (ex) mse_train:要素数4
    mse_test = np.zeros(k)  #要素数kの一次元配列で要素は0 (ex) mse_train:要素数4
    for i in range(0, k): # 0-i-(k-1) (ex) 0-i-3
        # 訓練データとテストデータに分割
        # (A) テストデータのインデックス
        i_test = np.fmod(range(n), k) #0~(n-1)をkで割った余りを格納 (ex) 0~15を4で割った余り
        x_test = x[i_test == i]   # テストデータ x 
        t_test = t[i_test == i]   # テストデータ t
        x_train = x[i_test != i]  # 訓練データ x
        t_train = t[i_test != i]  # 訓練データ t
        # wを訓練データで決める
        w = fit_gauss_func(x_train, t_train, m)
        # 訓練データのMSEを計算
        mse_train[i] = mse_gauss_func(x_train, t_train, w)
        # テストデータのMSEを計算
        mse_test[i] = mse_gauss_func(x_test, t_test, w)
    return mse_train, mse_test
```

```python
# リスト 5-4-(1)
import numpy as np
import matplotlib.pyplot as plt

# データのロード ----------
data = np.load("ch5_data.npz")
X = data["X"] # 年齢
X_min = 0
X_max = data["X_max"] # X_max=30
N = data["N"] # データ数16
T = data["T"] # 身長

# モデル A (w[0]に漸近していくモデル)----------
def model_A(x, w):
    y = w[0] - w[1] * np.exp(-w[2] * x)  # 式5-71
    return y

# モデル A 表示 ----------
def show_model_A(w):
    x = np.linspace(X_min, X_max, 100) # 0~30を100分割する
    y = model_A(x, w) # モデルの予測値を出す
    plt.plot(x, y, "gray", linewidth=4)

# モデル A の平均二乗誤差(MSE) ----------
def mse_model_A(w, x, t):
    y = model_A(x, w) # モデルの予測値を出す
    mse = np.mean((y - t) ** 2)  # 式5-72 平均二乗誤差(MSE)
    return mse
```

```python
# リスト 5-4-(2)
from scipy.optimize import minimize # パラメーターの極小値解を出力するライブラリ

# モデル A のパラメータ最適化  
def fit_model_A(w_init, x, t): # 最小値を求める関数、パラメーターの初期値を与える
    res = minimize(mse_model_A, w_init, args=(x, t), method="powell") # args=(x, t):最適化するパラメーター以外の引数を指定, method:方法
    return res.x    # 最適化したwの値を参照
```

```python
# リスト 5-4-(3)
# メイン ----------
w_init = np.array([100.0, 0.0, 0.0])  # wの初期値
w = fit_model_A(w_init, X, T)         # 最適化されたwを計算
sd = np.sqrt(mse_model_A(w, X, T))    # SD(平均二乗誤差)を計算
# 結果表示
print(f"w0 = {w[0]:.2f}, w1 = {w[1]:.2f}, w2 = {w[2]:.2f}")
print(f"SD = {sd:.2f} cm")

# グラフ描画 ----------
plt.figure(figsize=(4, 4))
show_model_A(w) # モデルの予測値をグラフに表示
plt.plot(       # 実測値をグラフに表示
    X, T, "cornflowerblue",
    marker="o", linestyle="None", markeredgecolor="black",
)
plt.xlim(X_min, X_max)  # 0~30
plt.grid()
plt.show()
```

## 5.7 モデルの選択

```python
# リスト 5-4-(4)
# 交差検証 モデル A ----------
def kfold_model_A(x, t, k):
    n = len(x) # (ex) n=16
    mse_train = np.zeros(k) # (ex)要素0の要素数16の一次元配列
    mse_test = np.zeros(k)  # (ex)要素0の要素数16の一次元配列
    for i in range(0, k): # 0-i-15
        # 訓練データとテストデータに分割
        i_test = np.fmod(range(n), k) # 0~15を15で割った余り
        x_test = x[i_test == i]
        t_test = t[i_test == i]
        x_train = x[i_test != i]
        t_train = t[i_test != i]
        # 精度を上げるため
        # リスト5-4-(3)で得た値を初期値に設定
        w_init = np.array([169.04, 113.66, 0.22]) # パラメーターの初期値
        w = fit_model_A(w_init, x_train, t_train) # 最適化されたwを計算
        mse_train[i] = mse_model_A(w, x_train, t_train) # 最適化されたwを使って平均二乗誤差を計算
        mse_test[i] = mse_model_A(w, x_test, t_test)    # 最適化されたwを使って平均二乗誤差を計算
    return mse_train, mse_test

# メイン ----------
# model A の交差検定
K = 16
Cv_A_train, Cv_A_test = kfold_model_A(X, T, K) # リーブワンアウト
sd_A_test = np.sqrt(np.mean(Cv_A_test)) # テスト用データの標準偏差
# 線形基底関数モデルの交差検定の結果のロード
sd_Gauss_test = np.load("ch5_Gauss_test.npy") # 基底関数の数ごとの平均二乗誤差(M=2~7)
# 結果のまとめ
SDs = np.append(sd_Gauss_test[0:5], sd_A_test) # np.append:新たな要素を追加
# 結果表示
print(f"Gauss(M=3) SD = {sd_Gauss_test[1]:.2f} cm")
print(f"Model A SD = {sd_A_test:.2f} cm")

# グラフ描画 ----------
M = range(6)
label = ["M=2", "M=3", "M=4", "M=5", "M=6", "Model A"]
plt.figure(figsize=(5, 3))
plt.bar(
    M, SDs,facecolor="cornflowerblue", align="center", 
    tick_label=label,
)
plt.show()
```
