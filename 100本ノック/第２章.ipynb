```python
#データを読み込む
import pandas as pd
uriage_data = pd.read_csv('/Users/shimizushota/python/100knock-data_analytics/2章/uriage.csv')
#uriage_data.describe()
#uriage_data.info()
print(uriage_data.head())

kokyaku_data = pd.read_excel('/Users/shimizushota/python/100knock-data_analytics/2章/kokyaku_daicho.xlsx')
print(kokyaku_data.head())
```

```python
#データの揺れを確認する
print(uriage_data['item_name'].head())
print(uriage_data['item_price'].head())

uriage_data['purchase_date'] = pd.to_datetime(uriage_data['purchase_date']) #データ型をdatetime64[ns]に変換することができる
uriage_data['purchase_month'] = uriage_data['purchase_date'].dt.strftime('%Y%m') #str=string,f=formatted dt=datetimeモジュール
res1 = uriage_data.pivot_table(index = 'purchase_month', columns = 'item_name', aggfunc = 'size', fill_value = 0) #月ごとの各item_nameの件数を集計
res1 #商品の数が26個のはずなのに99個表示されてしまっている

res2 = uriage_data.pivot_table(index = 'purchase_month', columns = 'item_name', values = 'item_price', aggfunc = 'sum', fill_value = 0)
res2
```

```python
#商品名の揺れを補正する
print(len(pd.unique(uriage_data['item_name']))) #item_nameが何種類あるのか確認

uriage_data['item_name'] = uriage_data['item_name'].str.upper() #小文字を大文字に変換
uriage_data['item_name'] = uriage_data['item_name'].str.replace('　', '') #全角のスペースを除去
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '') #半角のスペースを除去
uriage_data.sort_values(by=['item_name'], ascending = True) #item_nameを昇順で並び替えている

print(len(uriage_data['item_name'].unique())) #正しく補正されているか確認する
print(uriage_data['item_name'].unique()) #正しく補正されているか確認する
```

```python
#金額欠損値の補完を行う
uriage_data.isnull().any(axis = 0) #item_priceに欠損値が含まれている
flg_is_null = uriage_data['item_price'].isnull() #item_priceの欠損値をT/Fで一行ずつ確認
for trg in list(uriage_data.loc[flg_is_null, 'item_name'].unique()): #欠損値の存在する商品名を抽出(リストアップ)　例）商品A,商品S,商品Z
    price = uriage_data.loc[(~flg_is_null) & (uriage_data['item_name'] == trg), 'item_price'].max() #uriage_dataの欠損値のある商品の値段をpriceに代入
    uriage_data.loc[(flg_is_null) & (uriage_data['item_name'] == trg), 'item_price'] = price #uriage_dataの欠損値のある場所にpriceを代入
print(uriage_data.head())
uriage_data.isnull().any() #欠損値がなくなったか確認

for trg in list(uriage_data['item_name'].sort_values().unique()): #商品名のリストを作る
    print(trg + 'の最大額：' + str(uriage_data.loc[uriage_data['item_name'] == trg, 'item_price'].max()) #特定の商品名で最も価格が高いものを抽出
           + 'の最小額：' + str(uriage_data.loc[uriage_data['item_name'] == trg, 'item_price'].min(skipna = False))) #特定の商品名で最も価格が低いものを抽出
#skipna = FalseはNanデータが存在する場合、最小値はNanと表示される
```

```python
#顧客名の揺れを補正する
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace('　', '') #全角でスペースを入れる
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace(' ', '') #半角でスペースを入れる
print(kokyaku_data['顧客名'].head())

flg_is_serial = kokyaku_data['登録日'].astype('str').str.isdigit() #astype('str')で登録日を文字列に変換して、str.isdigit()で文字列が数字か判定している(T/F)
print(flg_is_serial.sum())

#数値から日付に変換　
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, '登録日'].astype('float') - 2, unit = 'D') + pd.to_datetime('1900/1/1')
fromSerial # Excelは1900/1/1から始まり、シリアル値は１なので０にするため−１する。また本来うるう年ではない1900/02/29が登録されているので−１する
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, '登録日']) #日付として取り込まれているデータも書式を揃える datetime型に変換
fromString
kokyaku_data['登録日'] = pd.concat([fromSerial, fromString])
print(kokyaku_data)

kokyaku_data['登録年月'] = kokyaku_data['登録日'].dt.strftime('%Y%m')
rslt = kokyaku_data.groupby('登録年月').count()['顧客名']
print(rslt)
print(len(kokyaku_data))

flg_is_serial = kokyaku_data['登録日'].astype('str').str.isdigit() #数値として入力されている値が０になったか確認
flg_is_serial.sum()
```

```python
#顧客名をキーに２つのデータを結合
join_data = pd.merge(uriage_data, kokyaku_data, left_on = 'customer_name', right_on ='顧客名' , how = 'left' )
joim_data = join_data.drop('customer_name', axis = 1)　#列を削除するのでaxis=1
join_data
```

```python
#クレンジングデータをダンプする
dump_data = join_data[['purchase_date', 'purchase_month', 'item_name', 'item_price', '顧客名', 'かな', '地域', 'メールアドレス', '登録日']]
dump_data

dump_data.to_csv('dump_data.csv', index = False)　#csvファイルの出力
```
